### Joining Tamara's file with new normalised file

```python
load_tamhara=pd.read_csv("/data/clientServices/clients/eBay_DE/speech/nlu/20190827_feature02_sent/Normalization/Norm_All/FirstResponses_WCNormalized",sep="\t",names=["uid","utterance","intent1","intent2"])
load_tamhara.head()
```

```python
joined_file=new_df.merge(load_tamhara,left_on="Filename",right_on="uid")
joined_file.head()
joined_file.loc[:,['Filename','Utterance',"Original Utterance","RU_Tag","utterance"]].head(100)
```

```python
file_comp=joined_file.loc[:,['Filename','Utterance',"Original Utterance","RU_Tag","utterance"]]
file_comp.columns=['Filename','Utterance_nil',"Original Utterance","RU_Tag","utterance_tamara"]
file_comp.head()
```

### Reading training file from MWB

```python
mwb_training= pd.read_csv("/data/users/nileshk/training_60k_MWB/Results/train_1",sep="\t",header=None,names=["filename_MWB","utterance_MWB","intent_MWB","intent2"])
mwb_training.head()
```
